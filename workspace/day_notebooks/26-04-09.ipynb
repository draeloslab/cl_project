```python
import cl
import pathlib
import matplotlib.pyplot as plt
import numpy as np
import time
```

```python
with cl.open() as neurons:
    recording = neurons.record()

    # Recording has started, place your application code here
    time.sleep(3)

    recording.stop()
```

```python
pathlib.Path(recording.file['path']).unlink()
```

```python
n_seconds = 4
n_channels = 64

intervals = []


neural_activity = []
with cl.open() as neurons:
    for tick in neurons.loop(ticks_per_second=300, stop_after_seconds=n_seconds):
        intervals.append([time.time()])
        spikes = np.zeros(n_channels)
        for spike in tick.analysis.spikes:
            spikes[spike.channel] += 1
        neural_activity.append(spikes)
        time.sleep(2/1000)
        intervals[-1].append(time.time())
neural_activity = np.array(neural_activity)
intervals = np.array(intervals)
```

```python
%matplotlib inline
intervals -= intervals.min()

fig, ax = plt.subplots()
ax.plot(intervals[:4].T, intervals[:4].T * 0 + 1);
```
